```python
def pseudoScalar(vec2a, vec2b):
    return vec2a[0] * vec2b[1] - vec2a[1] * vec2b[0]

def length(vec2):
    return (vec2[0] ** 2 + vec2[1] ** 2) ** 0.5

def createVec2(pointA, pointB):
    return (pointB[0] - pointA[0], pointB[1] - pointA[1])
```

### Задача: Вычисление площади многоугольника заданного координатами своих вершин.

```python
from IPython.display import Image
Image(filename='3_img.png')
```

```python
A = (1, 1)
B = (2, 4)
C = (3, 1)
D = (5, 5)
E = (6, 3)

def getArea(A1, A2, B1, B2):
    A1B1 = length(createVec2(A1, B1))
    A2B2 = length(createVec2(A2, B2))
    return 0.5 * (A1B1 + A2B2) * (B2[0] - B1[0])


points = [A, B, D, E, C]
bList = []
for point in points:
    bList.append((point[0], 0))

areaSum = 0

for i in range(len(points)):
    
    idx = i % len(points) - 1
    print("Temp areas = " + str(getArea(points[idx], points[idx + 1], bList[idx], bList[idx + 1])))
    areaSum += getArea(points[idx], points[idx + 1], bList[idx], bList[idx + 1])
areaSum
```
